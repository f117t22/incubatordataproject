```python
import pandas as pd
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('/Users/weibocheng/Desktop/Arrest_Data_from_2010_to_Present.csv')
df = pd.DataFrame(data)
```

```python
newdate = pd.to_datetime(df['Arrest Date'])
```

```python
df['newdate'] = newdate
sorteddf = df.sort_values(by='newdate')
datevsCrime = sorteddf['newdate'].value_counts()
datevsCrimesorted = datevsCrime.sort_index()
```

```python
# crime data overall trend
datevsCrimesorted.plot()
```

```python
plt.show()
```

```python
#crime data seasonality 
subindex = pd.date_range(start='1/1/2017', end='10/26/2019')
```

```python
subdatecrime = datevsCrimesorted[subindex]
```

```python
subdatecrime.plot()
```

```python
plt.show()
```
